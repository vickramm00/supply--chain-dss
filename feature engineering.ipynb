```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
PALETTE = ["#2563EB", "#16A34A", "#DC2626", "#F59E0B", "#7C3AED", "#0891B2"]
sns.set_theme(style="whitegrid", font_scale=1.1)
```

```python
df = pd.read_csv("bdspref2511.csv")
```

```python
df = df.dropna(subset=["cooalpha", "codalpha"])
print(f"\nRows after dropping nulls: {len(df):,}")
```

Now that `df` is properly loaded and cleaned, we can proceed with feature extraction.

```python
df["pref_claimed"] = df["use"].isin(["u10", "u11"]).astype(int)

print(f"\n  pref_claimed distribution:")
print(f"    Claimed (1)     : {df['pref_claimed'].sum():,}  ({df['pref_claimed'].mean()*100:.1f}%)的发展)")
print(f"    Not claimed (0) : {(df['pref_claimed']==0).sum():,}  ({(1-df['pref_claimed'].mean())*100:.1f}%)")
```

```python
df["value_per_kg"] = np.where(
    df["netmass"] > 0,
    df["statvalue"] / df["netmass"],
    0
)
print(f"\n  value_per_kg — mean: £{df['value_per_kg'].mean():,.2f}/kg")
```

```python
df["same_country"] = (df["cooalpha"] == df["codalpha"]).astype(int)
print(f"\n  same_country = 1 : {df['same_country'].sum():,} records ({df['same_country'].mean()*100:.1f}%)")
```

```python

df["is_eligible"] = df["eligibility"].isin(["e1", "e2"]).astype(int)
print(f"\n  is_eligible = 1  : {df['is_eligible'].sum():,} records ({df['is_eligible'].mean()*100:.1f}%)")

```

```python
df["log_statvalue"] = np.log1p(df["statvalue"])
print(f"\n  log_statvalue — mean: {df['log_statvalue'].mean():.3f}, std: {df['log_statvalue'].std():.3f}")
```

```python
df["log_statvalue"] = np.log1p(df["statvalue"])
print(f"\n  log_statvalue — mean: {df['log_statvalue'].mean():.3f}, std: {df['log_statvalue'].std():.3f}")
```

```python

print("\n  All new columns created:")
new_cols = ["pref_claimed", "value_per_kg", "same_country", "is_eligible", "log_statvalue", "hs_chapter"]
for c in new_cols:
    print(f"    ✓ {c}")
```
