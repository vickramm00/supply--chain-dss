```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python


df = pd.read_csv("bdspref2511.csv")
```

```python
print(f"  Rows : {len(df):,}")
print(f"  Cols : {len(df.columns)}")
```

```python
column_info = {
    "cooalpha"    : "Country of Origin (ISO 2-letter code, e.g. CN = China)",
    "codalpha"    : "Country of Destination (ISO 2-letter code, e.g. GB = UK)",
    "comcode"     : "8-digit Commodity Code (HS/CN trade classification)",
    "statreg"     : "Statistical Regime (1=normal, 2=supplementary, 3=special)",
    "eligibility" : "Preference Eligibility (e1=fully eligible, e2=partial, e3=conditional, e5=not eligible)",
    "use"         : "Preference Use — was it claimed? (u10=fully, u11=partially, u30=not claimed)",
    "perref"      : "Period Reference in YYYYMM format — 202511 = November 2025",
    "statvalue"   : "Statistical Trade Value in £ GBP",
    "netmass"     : "Net Mass of the shipment in kilograms",
    "suppunit"    : "Supplementary Unit (pieces/litres etc., varies by commodity)",
}

print("\n── COLUMN DESCRIPTIONS ──")
for col_name, description in column_info.items():
    print(f"  {col_name:<12}: {description}")
```

```python
print("\nData Types:")
print(df.dtypes)
```

```python
print("\nMissing (Null) Values per Column:")
print(df.isnull().sum())
```

```python
print("\nDuplicate rows:", df.duplicated().sum())
```

```python
df = df.dropna(subset=["cooalpha", "codalpha"])
print(f"\nRows after dropping nulls: {len(df):,}")
```

```python
print(df[["statvalue", "netmass", "suppunit"]].describe().round(2))
```

```python
print("\n── Categorical Field Distributions ──")
for col in ["statreg", "eligibility", "use"]:
    vc = df[col].value_counts()
    print(f"\n{col}:")
    for value, count in vc.items():
        pct = count / len(df) * 100
        print(f"  {value:<6} : {count:>7,}  ({pct:.1f}%)")
```
